### independent

$P(A \cap B) = P(A)P(B)$

### disjoint
$P(A \cap B)=0$  <br> 

$
\begin{align}
P(A \cup B) &= P(A)+P(B) -P(A\cap B) \\
  &= P(A)+P(B)-0 \\
  &= P(A)+P(B)
\end{align}
$
